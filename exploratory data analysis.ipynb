```python
## Skewness and Inter Quartile Range of the dataset
iqr_values = iqr(number_data, axis=0)
skewness_values = number_data.apply(skew, axis=0)

print("Interquartile Range (IQR) for Numeric Columns:")
iqrs = pd.DataFrame()
iqrs["Features"] = number_data.columns
iqrs["IQR"] = iqr_values
iqrs.reset_index(drop=True, inplace=True)
iqrs
```

```python
## barplot of the values of IQR column
iqrs = iqrs.sort_values(by=['IQR'], ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(iqrs['Features'], iqrs['IQR'])
plt.xlabel('Features')
plt.ylabel('IQR')
plt.xticks(rotation=90)
plt.show()
```

```python
## skewness
print("Skewness for Numeric Columns:")
skews = pd.DataFrame(skewness_values, columns=['Skewness'])
skews
```

```python
# Calculating escape velocities
G = 6.67430e-11  # Gravitational constant in m^3/kg/s^2
data['Escape Velocity'] = np.sqrt(2 * G * data['P_MASS'] / data['P_RADIUS'])
# Plot of estimated temperature vs. escape velocity
plt.figure(figsize=(10, 6))
plt.scatter(data['Escape Velocity'], data['S_TEMPERATURE'], alpha=0.5)
plt.xlabel('Escape Velocity (m/s)')
plt.ylabel('Estimated Temperature (K)')
plt.title('Estimated Temperature vs. Escape Velocity for Exoplanets')
plt.grid(True)
plt.show()
```

```python
# Calculating the average metallicities
data['S_METALLICITY_AVG'] = (data['S_METALLICITY'] + data['S_METALLICITY_ERROR_MIN'] + data['S_METALLICITY_ERROR_MAX']) / 3
data['S_METALLICITY_ERROR_AVG'] = (data['S_METALLICITY_ERROR_MIN'] + data['S_METALLICITY_ERROR_MAX']) / 2
# Calculating the correlation coefficient between host star ages and average metallicities
correlation_coefficient = data['S_AGE'].corr(data['S_METALLICITY_AVG'])
# Creating a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(data['S_AGE'], data['S_METALLICITY_AVG'], alpha=0.5)
plt.xlabel('Host Star Age')
plt.ylabel('Average Metallicities')
plt.title('Correlation between Host Star Age and Average Metallicities')
# Including correlation coefficient in the plot
plt.text(
    data['S_AGE'].min(),
    data['S_METALLICITY_AVG'].max(),
    f'Correlation Coefficient: {correlation_coefficient:.6f}',
    fontsize=12,
    color='red',
)
plt.grid(True)
plt.show()
# Printing the correlation coefficient
print(f'Correlation Coefficient: {correlation_coefficient:.6f}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['S_AGE'], np.log(data['S_METALLICITY']), alpha=0.5)
plt.xlabel('Host Star Age')
plt.ylabel('Host Star Metallicity')
plt.title('Correlation between Host Star Age and Metallicity')
plt.xlim(0, 10)
plt.ylim(-3, 0)
# Including correlation coefficient in the plot
plt.text(
    data['S_AGE'].min(),
    data['S_METALLICITY'].max(),
    f'Correlation Coefficient: {correlation_coefficient:.5f}',
    fontsize=12,
    color='red',
)
plt.grid(True)
plt.savefig('correlation_plot.png', dpi=300)
plt.show()
## Image file given in report
```

```python
## Feature Engineering
## Object Features
strcol = []
for i in data.columns:
    if data[i].dtype == "object":
        strcol.append(i)
```

```python
strcol
```

```python
## For Label Encoding the Object columns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lab = LabelEncoder()
ohe = OneHotEncoder()
```

```python
## Label Encoding the Object columns
for i in strcol:
    data[i] = lab.fit_transform(data[i])
```

```python
## columns having full null values
null = []
for i in data.columns:
    if data[i].isnull().sum() == len(data):
        null.append(i)
```

```python
null
```

```python
## Dropping the columns having full null values
data.drop(null, axis=1, inplace=True)
```

```python
## Imputation by K-Nearest Neighbors
## Giving most efficient performance
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
## saving the data
data.to_csv("data_processed_2.csv", index=False)
```

```python
## Imputing the nan values with mean (second option)
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(data)
data = pd.DataFrame(data=imp.transform(data), columns=data.columns)
```
